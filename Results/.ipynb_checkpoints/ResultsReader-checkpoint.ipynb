This function reads data from files based on the format outputted.

```python
from matplotlib import pyplot
import pandas as pd
import seaborn as sns

def read_results(filename):
    f = open((filename + ".txt"), "r")
    reading_data = True #Show that you are reading data. The first line is ignored
    cur_line = f.readline() #Get the first line
    data = []
    test_data = []
    while cur_line != "": #Whilst the end of the file has not been reached
        cur_line = f.readline() #Read the current line
        if reading_data == False: #If you are NOT reading data
            if cur_line != "": #Check that you haven't reached the end of the file
                if cur_line[0] == "!": #If the line begins with !, then the next lines are data to be read
                    reading_data = True
                    test_data = [] 
        else: #If you ARE reading data
            if cur_line[0] == "#": #If the line begins with #, then the next lines are not valid data
                reading_data = False #Mark that you are not reading the lines
                data.append(test_data)
            else: #A line of data is being read
                current_number = "" #Mark that there is no current number
                reading_fitness = False #Mark that the fitness is not yet being read
                for cur_char in cur_line: #For each character
                    if reading_fitness == False and cur_char == "[": #If you are not reading fitness and the character is [
                        reading_fitness = True #Note that this is the start of the fitness number
                    elif reading_fitness == True: #If you are reading fitness
                        if cur_char != "]" and cur_char != "\n": #And the current character is a number
                            current_number += cur_char #Add the number to the string
                test_data.append(float(current_number)) #Add the number to the current tests data
    f.close()
    return data

base_test = read_results("basetest")
base_testc = read_results("basetestc")
```

These two functions plot data based on a list, and a list of labels. The labels must be manually placed in.

```python
def convert_to_sns_data(input_list, label_list):
    data_dict = {}
    for index in range(len(label_list)): #For each label
        data_dict[label_list[index]] = input_list[index] #Add its corresponding list of results
    return pd.DataFrame(data=data_dict)

def draw_data_plot(input_name, label_list, c_test):
    input_list = read_results(input_name)
    if c_test:
        input_list.append(base_testc[0])
    else:
        input_list.append(base_test[0])
    data = convert_to_sns_data(input_list, label_list)
    sns.relplot(data=data, kind="line")
```

```python
draw_data_plot("muttest", [0.05, 0.2, 0.3, 0.1], False)
draw_data_plot("muttestc", [0.05, 0.2, 0.3, 0.1], True)
```

Displaying the best wind farm, as a result of mutation tests, at 0.3 with an AEP of 169.07854546312163

```python
import py_wake
from py_wake.site import UniformWeibullSite, UniformSite
from py_wake.examples.data.iea37 import IEA37_WindTurbines
from py_wake import NOJ

def setup_site():
    wind_turbines = IEA37_WindTurbines()
    site = UniformWeibullSite(p_wd = [1], a = [9.8], k = [2], ti = 0.1)
    noj = NOJ(site,wind_turbines)
    return noj

def divide_range(start, end, precision):
    point_list = []
    step = (end - start)/precision
    for i in range(precision):
        point_list.append(start + (step * i))
    return point_list
  
def create_grid_ranges(x_start, x_end, y_start, y_end, precision):
    grid_ranges = []
    x_list = divide_range(x_start, x_end, precision)
    y_list = divide_range(y_start, y_end, precision)
    for y_val in y_list:
        for x_val in x_list:
            grid_ranges.append([y_val, x_val])
    return grid_ranges
   
def get_turbine_locations(input_list, grid):
    x_positions = []
    y_positions = []
    for index in input_list:
        y_positions.append(grid[index][0])
        x_positions.append(grid[index][1])
    return x_positions, y_positions

SITE_CONFIG = setup_site()
SITE_POSITIONS = create_grid_ranges(0, 1000, 0, 1000, 1000)
best_layout = [926174, 918724, 619947, 7932, 989336, 174710, 211953, 505656, 21518, 947967, 419188, 15011, 176187, 891006, 27279, 522038]
best_x, best_y = get_turbine_locations(best_layout, SITE_POSITIONS)
result = SITE_CONFIG(best_x, best_y)

wind_turbines = IEA37_WindTurbines()
%matplotlib inline
pyplot.figure()
aep = result.aep()
wind_turbines.plot(best_x,best_y)
c =pyplot.scatter(best_x, best_y, c=aep.sum(['wd','ws']))
pyplot.colorbar(c, label='AEP [GWh]')
pyplot.title('AEP of each turbine')
pyplot.xlabel('x [m]')
pyplot.ylabel('[m]')
```
